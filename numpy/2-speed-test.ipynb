numpy's array is way faster than list (since it dont use references on other memory addresses)

```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
def list_view(size):
    start_time = time.time()
    l1 = range(size)
    l2 = range(size)
    summation = [ l1[i] + l2[i] for i in range(len(l1)) ]
    return time.time() - start_time

def numpy_view(size):
    start_time = time.time()
    n1 = np.arange(size)
    n2 = np.arange(size)
    summation = n1 + n2
    return time.time() - start_time

```

```python
size = 1000000
lt = list_view(size)
nt = numpy_view(size)
print('list view execution time: ', lt)
print('numpy view execution time: ', nt)
print(f'numpy is {lt / nt} times faster')
```

lets check it for different sizes

```python
sizes = np.arange(10000, 1000001, 5000)
plt.figure(figsize=(16, 9))
plt.plot(sizes, [list_view(_size) for _size in sizes], label='list')
plt.plot(sizes, [numpy_view(_size) for _size in sizes], label='numpy')
plt.legend()
plt.ylabel('time')
plt.xlabel('size')
plt.show()
```
