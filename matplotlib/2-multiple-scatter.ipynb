# Multiple datasets in one scatter

we can easily declare multiple plt.scatter then define a label and use .legend() method to show the labels on the board

```python
import matplotlib.pyplot as plt


# Sample data
x_data1 = [1, 2, 3, 4, 5]
y_data1 = [10, 12, 15, 18, 20]

x_data2 = [2, 3, 4, 5, 6]
y_data2 = [8, 11, 14, 16, 19]

# Plotting the first set of data
plt.scatter(x_data1, y_data1, label='Data Set 1', color='blue')

# Plotting the second set of data
plt.scatter(x_data2, y_data2, label='Data Set 2', color='red')

# Adding labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Two Data Sets')
plt.legend()

plt.show()
```
