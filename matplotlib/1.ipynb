### we can create different types of graphs, here are some type example:

- 'o' which put circles
- '-' which is regular line
- '*' 
- '.' clearly this one is just dots
- '^' triangles



```python
plt.plot([1, 2, 3], [1, 4, 9], 'o')
```

```python
plt.plot([1, 2, 3], [1, 4, 9], 'ro')
```

this 'r' or 'y' etc, are the colors

```python
plt.plot([1, 2, 3], [1, 4, 9], 'yo')
```

```python
plt.plot([1, 2, 3], [1, 4, 9], 'r+')
```

```python
plt.plot([1, 2, 3], [1, 4, 9], '*')
```

### we can separate the joints with using another style for the graph

for example '-' is a unified line, we can use '--' instead to make it separate joints and use - for both of them


```python
plt.plot([1, 2, 3], [1, 4, 9], 'r--')
```

```python
a = np.arange(0, 5, .2)
a
```

```python
plt.plot(a, a**2)
```

we can even use different types for the joints :

```python
plt.plot(a, a, 'r-.')
```

however '-.' is not equal with '.-'

```python
plt.plot(a, a, 'r.-')
```

```python
plt.plot(a, a, 'r-.', a, a**2, 'bs-')
```

```python
plt.plot(a, a, 'r-.', a, a**2, 'bs-',a, a **3, 'g^')
```
